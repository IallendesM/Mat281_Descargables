```python
#importacion de modelos y otras cosas
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,recall_score, precision_score,f1_score
from sklearn.metrics import roc_curve, auc
import time
```

```python
X = digits.drop(columns="target").values
y = digits["target"].values
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=34,test_size=0.1)
#como los datos representan una escala de grises, no los voy a escalar
print('largo del train set:',len(X_train))
print('largo del test set:',len(X_test))
```

```python
#modelos
lr = LogisticRegression()
knn = KNeighborsClassifier()
rf = RandomForestClassifier()


#entrenamiento
start = time.time()
lr.fit(X_train,y_train)
end = time.time()
lr_time=end-start #en segundos
start = time.time()
knn.fit(X_train,y_train)
end = time.time()
knn_time=end-start #en segundos
start = time.time()
rf.fit(X_train,y_train)
end = time.time()
rf_time=end-start #en segundos
#no es la idea, pero voy a usar el test set como validación
#si los scores son parecidos, voy a asumir que no hay overfitting
print('Logistic Regression:')
print('train score:',lr.score(X_train,y_train))
print('test score:',lr.score(X_test,y_test))
print('tiempo:',lr_time,'[s]')

print('K Nearest Neighbors:')
print('train score:',knn.score(X_train,y_train))
print('test score:',knn.score(X_test,y_test))
print('tiempo:',knn_time,'[s]')

print('Random Forest:')
print('train score:',rf.score(X_train,y_train))
print('test score:',rf.score(X_test,y_test))
print('tiempo:',rf_time,'[s]')
```

```python
def classification_metrics(df):
    """
    Aplicar las distintas métricas definidas
    :param df: dataframe con las columnas: ['y', 'yhat']
    :return: dataframe con las métricas especificadas
    """
    y_true = df['y']
    y_pred = df['yhat']

    accuracy = round(accuracy_score(y_true, y_pred), 4)
    recall = round(recall_score(y_true, y_pred, average='macro'), 4)
    precision = round(precision_score(y_true, y_pred, average='macro'), 4)
    fscore = round(f1_score(y_true, y_pred, average='macro'), 4)

    df_result = pd.DataFrame({'accuracy': [accuracy],
                              'recall': [recall],
                              'precision': [precision],
                              'fscore': [fscore]})

    return df_result

def evaluate_model(model, X_test, y_test):
  y_pred = model.predict(X_test)
  print(confusion_matrix(y_test,y_pred))
  metrics_df = classification_metrics(pd.Dataframe({'y': y_test, 'yhat':y_pred}))
  print(metrics_df)
```

```python
y_lr = lr.predict(X_train)
y_knn = knn.predict(X_train)
y_rf = rf.predict(X_train)
print('Logistic Regression:')
print(classification_report(y_train,y_lr))
print(confusion_matrix(y_train,y_lr))
print('K Nearest Neighbors:')
print(classification_report(y_train,y_knn))
print(confusion_matrix(y_train,y_knn))
print('Random Forest:')
print(classification_report(y_train,y_rf))
print(confusion_matrix(y_train,y_rf))
```

* ¿Cuál modelo es mejor basado en sus métricas?

R: Logistic Regression o Random Forest, tienen metricas perfectas en el **train set**.
* ¿Cuál modelo demora menos tiempo en ajustarse?


R: K Nearest Neighbors con ~0.001[s]
* ¿Qué modelo escoges?


R: K Nearest Neighbors, ya que, deberia de ser el más robusto en este caso, debido a que la diferencia entre su score en el train y en el test es de 0.01, mientras que el de Logistic es 0.04 y el de Random Forest 0.03. Por lo tanto, es el que menos sufre de overfitting (Para un mejor analisis de overfitting deberiamos hacer crossvalidation para estar más seguros)

__Comprensión del modelo:__ Tomando en cuenta el **mejor modelo entontrado** en el `Ejercicio 3`, debe comprender e interpretar minuciosamente los resultados y gráficos asocados al modelo en estudio, para ello debe resolver los siguientes puntos:


 * **Curva AUC–ROC**: Replica el ejemplo del siguiente  [link](https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html#sphx-glr-auto-examples-model-selection-plot-roc-py) pero con el modelo, parámetros y métrica adecuada. Saque conclusiones del gráfico.

```python
# Obtener las probabilidades de predicción para cada clase
y_pred_proba = knn.predict_proba(X_train)

# Calcular la curva ROC y el puntaje AUC-ROC para cada clase
n_classes = len(knn.classes_)
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_train, y_pred_proba[:, i], pos_label=i)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Graficar las curvas AUC-ROC para cada clase
plt.figure(figsize=(8, 6))
colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2, label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas AUC-ROC para el modelo de KNN')
plt.legend(loc="lower right")
plt.show()
```

R: Concluimos que el clasificador es muy bueno. Pues los AUC scores son de ~1 para todas las clases. Esto se puede deber a que hay pocas muestras donde se equivoque.

### Ejercicio 5


__Visualizando Resultados:__ A continuación se provee código para comparar las etiquetas predichas vs las etiquetas reales del conjunto de _test_.

```python
def mostrar_resultados(digits, model, nx=5, ny=5, label="correctos"):
    """
    Muestra los resultados de las predicciones de un modelo de clasificación en particular.
    Se toman aleatoriamente los valores de los resultados.

    - label == 'correctos': muestra los valores en los que el modelo acierta.
    - label == 'incorrectos': muestra los valores en los que el modelo no acierta.

    Observación: El modelo que se recibe como argumento no debe estar entrenado.

    :param digits: dataset 'digits'
    :param model: modelo de sklearn
    :param nx: número de filas (subplots)
    :param ny: número de columnas (subplots)
    :param label: 'correctos' o 'incorrectos'
    :return: gráficos matplotlib
    """

    X = digits.drop(columns="target").values
    y = digits["target"].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    model.fit(X_train, y_train)  # Ajustar el modelo
    y_pred = model.predict(X_test)

    # Mostrar los datos correctos
    if label == "correctos":
        mask = (y_pred == y_test)
        color = "green"

    # Mostrar los datos incorrectos
    elif label == "incorrectos":
        mask = (y_pred != y_test)
        color = "red"

    else:
        raise ValueError("Valor incorrecto")

    X_aux = X_test[mask]
    y_aux_true = y_test[mask]
    y_aux_pred = y_pred[mask]

    # Mostrar los resultados
    n_samples = min(nx * ny, len(X_aux))
    indices = np.random.choice(len(X_aux), n_samples, replace=False)
    fig, ax = plt.subplots(nx, ny, figsize=(12, 12))

    for i, index in enumerate(indices):
        data = X_aux[index, :].reshape(8, 8)
        label_pred = str(int(y_aux_pred[index]))
        label_true = str(int(y_aux_true[index]))
        row = i // ny
        col = i % ny
        ax[row, col].imshow(data, interpolation='nearest', cmap='gray_r')
        ax[row, col].text(0, 0, label_pred, horizontalalignment='center', verticalalignment='center', fontsize=10, color=color)
        ax[row, col].text(7, 0, label_true, horizontalalignment='center', verticalalignment='center', fontsize=10, color='blue')
        ax[row, col].get_xaxis().set_visible(False)
        ax[row, col].get_yaxis().set_visible(False)

    plt.show()
```
